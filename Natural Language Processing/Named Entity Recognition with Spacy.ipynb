# Named Entity Recognition with Spacy

```python
import spacy
```

```python
model = spacy.load("en_core_web_sm")
```

```python
with open("data/wiki_us.txt") as f:
    txt = f.read()
```

```python
print(txt)
```

```python
doc = model(txt)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
```

```python
displacy.render(doc, style="ent")
```
